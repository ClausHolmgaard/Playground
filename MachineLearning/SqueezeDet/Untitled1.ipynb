```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bce(y, y_hat):
    return y * (-np.log(y_hat + 1e-16)) + (1-y) * (-np.log(1-y_hat + 1e-16))
```

```python
y = np.linspace(1.0, 1.0, num=1000)
y_hat = np.linspace(0, 1.0, num=1000)
```

```python
b = bce(y, y_hat)
```

```python
f = plt.figure(figsize=(15, 8))
plt.plot(y_hat, b)
plt.title("Binary Crossentropy as prediction approaches ground truth")
plt.show()
```
