```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
num_data = 100
a = 2
b = 30
noise_scale = 20
```

```python
x = np.linspace(-100, 100, num=num_data)
y = a*x + b
plt.scatter(x, y)
```

```python
noise = np.random.normal(size=num_data)
y = y + noise_scale * noise
plt.scatter(x, y)
```

```python
prev_ind = 0
step = 10
plot_cols = 3

num_slices = np.ceil(num_data / step)
plot_rows = int(num_slices/plot_cols) + 1
f, subs = plt.subplots(plot_cols, plot_rows, figsize=(15, 4*plot_rows))
subs = subs.ravel()

counter = 0
for i in range(step, num_data + 1, step):
    x_slice = x[prev_ind:i]
    y_slice = y[prev_ind:i]
    prev_ind = i
    
    reg = LinearRegression().fit(x_slice.reshape(-1, 1), y_slice.reshape(-1, 1))
    a_predict = reg.coef_[0][0]
    b_predict = reg.intercept_[0]
    y_predict = reg.predict(x.reshape(-1, 1))

    subs[counter].set_title(f"iteration: {counter}  a: {a_predict:.2f}, b: {b_predict:.2f}")
    subs[counter].scatter(x, y_predict)
    subs[counter].scatter(x, y, c='green')
    
    counter += 1
```
